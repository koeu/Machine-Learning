## 상관관계 시각화

```python
%matplotlib inline
import pandas as pd
import seaborn as sns

df = pd.read_csv('./exam.csv')

sns.heatmap(df.corr())
```

색이 밝을수록 상관 계수가 더 높다는 의미입니다.  
읽기 점수(reading score)와 쓰기 점수(writing score) 사이의 상관 관계가 가장 강하다는 것을 한 눈에 알 수 있네요.

annot=True 옵션을 추가해주면, 색상 뿐 아니라 숫자도 함께 보여줍니다.

```python
sns.heatmap(df.corr(), annot=True)
```

```python
import pandas as pd
import seaborn as sns
```

```python
body_df = pd.read_csv('./body.csv', index_col = 0)
body_df.head()
```

```python
body_df['Height']
```

```python
body_df['Height'].value_counts()
```

키 순으로 정렬 (index 순으로 정렬)

```python
body_df['Height'].value_counts().sort_index()
```

```python
body_df['Height'].value_counts().sort_index().plot()
```

매끄러운 곡선으로 만들기 - seaborn 사용

```python
sns.kdeplot(body_df['Height'])
```

추측을 어느정도로 할지 설정 : bw (히스토그램에서 bin 과 같은 역할)

```python
sns.kdeplot(body_df['Height'], bw = 0.5)
```

bw 가 너무 커지면 부정확해짐

```python
sns.kdeplot(body_df['Height'], bw = 2)
```

## KDE 활용 예시

1) histogram

```python
body_df.plot(kind = 'hist', y = 'Height', bins = 15)
```

histogram 위에 KDE plot 겹치기

```python
sns.distplot(body_df['Height'], bins = 15)  
```

2) box plot

```python
body_df.plot(kind = 'box', y = 'Height')  #통계 요약본
```

```python
sns.violinplot(y=body_df['Height'])  #분포 자체를 보여줌
```

3) 키와 몸무게의 상관도 - scatter

```python
body_df.plot(kind = 'scatter', x='Height', y='Weight')
```

등고선 - 높낮이를 표현
x= 키의 kde, y=몸무게의 kde 를 3차원으로
선이 가까울수록 가파름, 선이 멀수록 평평함

```python
sns.kdeplot(body_df['Height'], body_df['Weight'])
```

```python
sns.kdeplot(body_df['Height'])
```

```python
sns.kdeplot(body_df['Weight'])
```

### LM plot

```python
body_df.plot(kind = 'scatter', x='Height', y='Weight')
```

regression line 그려줌 - 키에 따른 몸무게 예측 가능

```python
sns.lmplot(data=body_df, x='Height', y='Weight')
```

## 카테고리별 시각화

```python
laptops_df = pd.read_csv('./laptops.csv')
laptops_df.head()
```

```python
laptops_df['os'].unique()
```

각 os가 어떤 특징을 갖는지 
ex) os 별 가격

#### 카테고리 별 그래프화 - "catplot"

```python
sns.catplot(data=laptops_df, x='os', y='price', kind='box')
```

```python
sns.catplot(data=laptops_df, x='os', y='price', kind='violin')
```

box, violin -> 각 os data 갯수를 모르니깐 그래프 신빙성 낮음 --> strip plot 사용

```python
sns.catplot(data=laptops_df, x='os', y='price', kind='strip')
```

```python
laptops_df['processor_brand'].unique()
```

프로세스별
--> hue = 색 : 프로세스에 따라 색깔 다르게 표시

```python
sns.catplot(data=laptops_df, x='os', y='price', kind='strip', hue='processor_brand')
```

점이 많이 있으면 보기 힘듬 -> swarm plot 사용 : 점들이 옆으로 펼쳐짐 

```python
sns.catplot(data=laptops_df, x='os', y='price', kind='swarm', hue='processor_brand')
```
